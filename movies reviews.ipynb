```python
import pandas as pd
from textblob import TextBlob 
import re, os
```

```python
DATA_PATH = "D:\data\movies_reviews"
file = pd.read_csv(os.path.join(DATA_PATH, "test.tsv"), sep='\t')
```

```python
def infrange(x, l, r):
    return x >=l and x < r
```

```python
def sentiment(phrase):
    LEFT = -1.0
    RIGHT = 1.0
    analysis = TextBlob(phrase)
    score = analysis.sentiment.polarity
    print(f"Popuarity: {score}")
    score_nom = (score - LEFT)/(RIGHT - LEFT) * 5
    return int(score_nom)
```

```python
sent = []
for index, row in file.iterrows():
    print(f"Processing line {index}")
    sent.append(sentiment(row["Phrase"]))
```

```python
res = pd.DataFrame({"PhraseId": file["PhraseId"], "Sentiment": sent})
```

```python
res.to_csv(os.path.join(DATA_PATH, "submit.csv"), index=False)
```
